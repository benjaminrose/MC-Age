# HR comparison
 
 There seems to be a differance between the Gupta and Campbell data set that might be from the cosmology or the increasted in quality of processing of the newer data set. These plots try compare the two data sets directly.

```python
# Campbell objects
# Campbell uncorrected distances
campbell_data = pd.read_csv('../data/CampbellHoltzman.tsv', sep='\t', usecols=['SNID', 'hr', 'err_mu'], index_col='SNID')
campbell_data.rename(columns={'hr': 'c_hr', 'err_mu': 'c_hr_uncert'}, inplace=True)
print(len(campbell_data))
campbell_data.head(5)
```

```python
# Gupta objects
# Guptas cosmology
gupta_data = pd.read_csv('../data/Gupta11_table2.tsv', sep='\t', skiprows=2, usecols=['x_1 uncert', 'HR'])#, index_col='SN ID')
gupta_data.rename(columns={'x_1 uncert': 'g_hr', 'HR': 'g_hr_uncert' }, inplace=True) # 'cause read_csv is shifting the headers over by one?
print(len(gupta_data))
gupta_data.head(5)
```

```python
 result = pd.concat([campbell_data, gupta_data], axis=1, join='inner')
```
